# Formulas estadisticas python

* x.count( ) =>> Contar
* x.size( ) =>> TamaÃ±o 
* x.mean( ) =>> Proemdio
* x.median( ) =>> Mediana 
* x.mode( ) =>> Moda
* x.std( ) =>> Desviacion tipica
* x.var( ) =>> Varianza
* x.quantile( ) =>> Cuartile
* x.min( ) =>> minimo
* x.max( ) =>> Maximo
* x.describe( ) =>> Descritivo
* x.cumsum( ) =>> suma acumulada
* x.cummax( ) =>> max acumulada
* x.cummin( ) =>> min acumulada
* x.cumprod( ) =>> productoria

```python
def Descriptivo(x):
    media=[]
    mediana=[]
    varianza=[]
    desviacion=[]
    maximo=[]
    minimo=[]
    contar=[]
    size=[]
    mode=[]
    sum_aco=[]
    max_aco=[]
    min_aco=[]
    prod=[]
    media.append(x.mean())
    mediana.append(x.quantile(0.5))
    varianza.append(x.var())
    desviacion.append(x.std())
    maximo.append(x.max())
    minimo.append(x.min())
    contar.append(x.count())
    mode.append(x.mode())
    sum_aco.append(x.cumsum())
    max_aco.append(x.cummax())
    min_aco.append(x.cummin())
    prod.append(x.cumprod())
    resultados = pd.DataFrame({'Media':media , 'Mediana':mediana,
                                'Varianza':varianza, 'Desviacion':desviacion,
                                  'Maximo':maximo, 'Minimo':minimo, 'Contar':contar,
                                  'Moda':mode, 'Suma_acomulada':sum_aco,
                                  'Maximo_acomulado':max_aco,'Minimo_acomulado':min_aco,
                                  'Productoria':prod})
    return resultados
       
```

```python
def Descriptivo2(x):
    variable =[]
    media=[]
    mediana=[]
    varianza=[]
    desviacion=[]
    maximo=[]
    minimo=[]

    for i in range(len(x.columns)):
        variable.append(x.columns[i])
        media.append(x[x.columns[i]].mean())
        mediana.append(x[x.columns[i]].quantile(0.5))
        varianza.append(x[x.columns[i]].var())
        desviacion.append(x[x.columns[i]].std())
        maximo.append(x[x.columns[i]].max())
        minimo.append(x[x.columns[i]].min())
        resultados = pd.DataFrame({'Variables':variable,'Media':media , 'Mediana':mediana,
                                'Varianza':varianza, 'Desviacion':desviacion,
                                  'Maximo':maximo, 'Minimo':minimo})
    return resultados
       
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, auc, confusion_matrix, roc_auc_score, roc_curve
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
df.columns
```

```python
fx = df[['GENDER', 'MONTH', 'DAY', 'HOUR', 'AMOUNT', 'Class']]
```

```python
cat = pd.get_dummies(fx[['GENDER', 'MONTH']], dtype=float)
cat.head()
```

```python
fz = pd.concat([fx[['DAY', 'HOUR', 'AMOUNT', 'Class']], cat], axis=1)
fz.head()
```

```python
x = fz.drop('Class', axis=1)
y = fz['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.75, random_state=123)
```

```python
rf_cl = RandomForestClassifier(n_estimators=100)
```

```python
rf_cl.fit(x_train, y_train)
```

```python
y_pred_rfcl = rf_cl.predict(x_test)
```

```python
conf_matrix = confusion_matrix(y_test, y_pred_rfcl)
conf_matrix
```

```python
import seaborn as sns

sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', cbar=False, 
            xticklabels=['No fraude 0', 'Fraude 1'], yticklabels=['No fraude 0', 'Fraude 1'])
plt.xlabel('Predicciones')
plt.ylabel('Valores reales')
plt.title('Matriz de ConfusiÃ³n')
plt.show()
```

```python
y_prob = rf_cl.predict_proba(x_test)[:, 1]
y_prob

# Calcular la curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Calcular la curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

roc_auc = auc(fpr, tpr)

# Visualizar la curva ROC

plt.figure(figsize=(5, 5))
plt.plot(fpr, tpr, color ='darkorange', lw = 2, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color = 'navy', lw = 2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()
```
